```python
import pandas as pd
```

```python
import requests
import json

#로스트아크 api에서 발급받은 키를 mytoken에 저장.
mytoken=''
token='bearer '+mytoken

headers={
    'accept' : 'application/json',
    'authorization' : token
}
```

```python
url = 'https://developer-lostark.game.onstove.com/news/events'
response=requests.get(url, headers=headers)
jsonobject=response.json()

print(response)
for list in jsonobject:
    print(list.get("Title"))
```

```python
name='초신성은하'
character_url='https://developer-lostark.game.onstove.com/characters/'+name+'/siblings'
response_character=requests.get(character_url, headers=headers)
allcharacters=response_character.json()
```
